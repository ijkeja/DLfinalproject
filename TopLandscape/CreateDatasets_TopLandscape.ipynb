The code below was used to create the reduced Top dataset. The created files were downloaded and stored in my Google Drive. In the model notebooks these modified datafiles are imported using gdown, so there is no need to run this code.

```python
# Import necessary modules
import pandas as pd
```

```python
# Download all three HDF5 files from Zenodo - https://zenodo.org/records/2603256

!wget "https://zenodo.org/record/2603256/files/train.h5?download=1" -O train.h5
!wget "https://zenodo.org/record/2603256/files/val.h5?download=1" -O val.h5
!wget "https://zenodo.org/record/2603256/files/test.h5?download=1" -O test.h5
```

# Training set

```python
# Get data and take 12k samples randomly and store it in a new file to Google Drive
df = pd.read_hdf('train.h5', key='table')
df = df.sample(n=12000, random_state=42)

save_path = '/content/TopLandscape_train_12k.csv'
df.to_csv(save_path, index=False)
```

# Validation set

```python
# Get data and take 4k samples randomly and store it in a new file to Google Drive
df = pd.read_hdf('val.h5', key='table')
df = df.sample(n=4000, random_state=42)

save_path = '/content/TopLandscape_val_4k.csv'
df.to_csv(save_path, index=False)
```

# Test set

```python
# Get data and take 4k samples randomnly and store it in a new file to Google Drive
df = pd.read_hdf('test.h5', key='table')
df = df.sample(n=4000, random_state=42)

save_path = '/content/TopLandscape_test_4k.csv'
df.to_csv(save_path, index=False)
```
